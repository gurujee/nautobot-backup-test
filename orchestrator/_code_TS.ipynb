```python
from dotenv import load_dotenv
import os
import requests
import json

load_dotenv()

url = os.getenv("NAUTOBOT_URL")
token = os.getenv("NAUTOBOT_TOKEN")

headers = {
    "Authorization": f"Token {token}",
    "Content-Type": "application/json"
}

response = requests.get(
    f"{url}/api/dcim/devices/",
    headers=headers,
    verify=False
)

print(f"STATUS CODE = {response.status_code}")

data = response.json()

print(json.dumps(data, indent=4))
```

```python
"""
Generic REST API helper functions.
"""

from typing import Dict
import requests

def api_get(url: str, headers: Dict) -> Dict:
    """
    Execute HTTP GET request.

    Args:
        url (str):
            Target API URL.

        headers (Dict):
            HTTP headers.

    Returns:
        Dict:
            JSON response body.
    """

    data = requests.get(
        url=url,
        headers=headers,
        verify=False,
        timeout=10,
    )

    data.raise_for_status()

    return data.json()

print (url)
```
